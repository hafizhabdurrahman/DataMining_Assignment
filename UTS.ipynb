## Training

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from skimage.io import imread, imshow
```

```python
#df = pd.read_csv('/content/drive/MyDrive/data3scene.csv')
df = pd.read_csv('/content/drive/MyDrive/data3scene.csv')
```

```python
df.head()
```

```python
x = df.iloc[:, :196608].values
y = df.iloc[:, 196608].values
```

```python
#Normalisasi
x = x / 255.0
```

```python
x = pd.DataFrame(x)
x.head()
```

```python
y = pd.DataFrame(y)
y.head()
```

```python
#Reshape -> 3D
x = x.values.reshape(-1,256,256,3)
```

```python
#Encode label
from sklearn.preprocessing import LabelEncoder
labelencoder_X_1 = LabelEncoder()
y = labelencoder_X_1.fit_transform(y)
```

```python
y = pd.DataFrame(y)
y.head()
```

```python
#Encode label
from keras.utils.np_utils import to_categorical

y = to_categorical(y, num_classes = 3)
```

```python
# Split dataset
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 0)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout
# Inisialisasi ANN
cl = Sequential()
```

```python
#first hidden layer
cl.add(Dense(128,
                     kernel_initializer='uniform',
                     activation='relu',
                     input_shape=(256,256,3)))
# Adding dropout to prevent overfitting
cl.add(Dropout(0.1))
```

```python
from keras.layers import Flatten
cl.add(Flatten())
```

```python
#cl.add(Dense(512,
#                     kernel_initializer='uniform',
#                     activation='relu'))
#cl.add(Dropout(0.1))
```

```python
#cl.add(Dense(128,
#                     kernel_initializer='uniform',
#                     activation='relu'))
#cl.add(Dropout(0.1))
```

```python
cl.add(Dense(16,
                     kernel_initializer='uniform',
                     activation='relu'))
cl.add(Dropout(0.1))
```

```python
cl.add(Dense(8,
                     kernel_initializer='uniform',
                     activation='relu'))
cl.add(Dropout(0.1))
```

```python
cl.add(Dense(3,
                     kernel_initializer='uniform',
                     activation='softmax'))
#clf1.add(Dropout(0.1))
```

```python
#clf1.add(Dense(1,
#                     kernel_initializer='uniform',
#                     activation='softmax'))
```

```python
# Compiling the ANN
cl.compile(loss='categorical_crossentropy',
             optimizer='adam',
             metrics=['accuracy'])
```

```python
cl.fit(X_train, y_train,
         epochs=20,
         batch_size=30,
         shuffle=True)
```

```python
pred = cl.predict(X_test)
pred_classes = np.argmax(pred,axis = 1) 
```

```python
y_test_classes = np.argmax(y_test,axis = 1)
```

```python
from sklearn.metrics import confusion_matrix
import itertools
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]

    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i, j],
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
  
```

## Hasil

```python
confusion_mtx = confusion_matrix(y_test_classes, pred_classes)
plot_confusion_matrix(confusion_mtx, classes = range(3))
```

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(y_test_classes,pred_classes)
print("Accuracy =", acc)
```
